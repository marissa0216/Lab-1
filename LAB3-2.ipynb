# Title
Digestion of fat with pancreatic lipase and bile salts.

# Purpose
To examine some aspects of the action of pancreatic lipase and bile salts on lipids.
This was done by recording the pH, as seen in the color change of litmus powder, and smell of
each test tube. Pancreatin contains the enzyme lipase, which is found in the pancreas and
breaks down lipid molecules at the optimum pH of 7.8 and optimum temperature of 37. Bile salts
are produced in the liver and act as an emulsifier by breaking ℃ lipids in the heavy cream into
smaller molecules thus making it easier for pancreatic lipase to break down the lipid molecules.

# Procedure
1. Add just enough litmus powder to a container of dairy cream to produce a medium bluecolor. Pour 3 ml of the litmus cream into 4 separate test tubes. Into two additional
testtubes pour 3 ml of 2% pancreatin. Preincubate the litmus cream and the pancreatin separately in a 37C water bath for 5 minutes.Then prepare tubes as follows:

  Tube #1: 3ml cream + 3 ml pancreatin
  Tube #2: 3 ml cream + 3 ml distilled water
  Tube #3: 3 ml + 3 ml pancreatin + pinch of bile salts
  Tube #4: 3 ml cream + 3 ml distilled water + pinch of bile salts.

2. Gently shake each tube for 30 seconds to mix in the bile salts. Incubate all four tubes in a 37C water bath for 1
hour checking every minute for the first 5 minutes or until the first tube changes color,then every 15 minutes for the rest of the hour. Record the time and number of the tube. Continue checking for the remainder of the

3. Remove the tubes from the water bath. Test the pH of each tube using pH paper and note the odor and color of each tube. NOTE: Blue litmus will turn pink in an acid environment.
